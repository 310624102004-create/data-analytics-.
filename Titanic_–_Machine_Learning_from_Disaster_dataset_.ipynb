## Feature Engineering

### Subtask:
Create new features: 'FamilySize' by combining 'SibSp' and 'Parch' plus one, an 'IsAlone' feature based on 'FamilySize', and 'Age groups' (e.g., child, adult, senior) by binning the 'Age' column.


**Reasoning**:
I need to implement the feature engineering steps as described in the instructions: create 'FamilySize', 'IsAlone', and 'AgeGroup' columns, then display the head of the DataFrame and value counts for the newly created features to verify their correctness.



```python
data_df['FamilySize'] = data_df['sibsp'] + data_df['Parch'] + 1

data_df['IsAlone'] = (data_df['FamilySize'] == 1).astype(int)

bins = [0, 12, 25, 60, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Senior']
data_df['AgeGroup'] = pd.cut(data_df['Age'], bins=bins, labels=labels, right=False)

print("DataFrame head with new features:")
print(data_df.head())

print("\nValue counts for FamilySize:")
print(data_df['FamilySize'].value_counts().sort_index())

print("\nValue counts for IsAlone:")
print(data_df['IsAlone'].value_counts())

print("\nValue counts for AgeGroup:")
print(data_df['AgeGroup'].value_counts().sort_index())

print("Feature engineering completed successfully.")
```

Train and compare various classification models on the training data, including Logistic Regression, Random Forest Classifier, Gradient Boosting Classifier, and XGBoost Classifier.


**Reasoning**:

## Model Evaluation

### Subtask:
Evaluate the performance of each trained model using key metrics: calculate accuracy scores on the validation set, generate confusion matrices to analyze prediction errors, and perform cross-validation to assess model robustness and generalization capabilities.


**Reasoning**:
I need to evaluate each trained model by making predictions on the validation set, calculating accuracy, generating confusion matrices, and performing 5-fold cross-validation, then printing these results for each model.



```python
models = {
    "Logistic Regression": lr_model,
    "Random Forest": rf_model,
    "Gradient Boosting": gb_model,
    "XGBoost": xgb_model
}

results = {}

print("\n--- Model Evaluation ---")

for name, model in models.items():
    print(f"\nEvaluating {name}...")

    # 1a. Make predictions on the validation set
    y_pred = model.predict(X_val)

    # 1b. Calculate and print accuracy score
    accuracy = accuracy_score(y_val, y_pred)
    print(f"Accuracy on validation set: {accuracy:.4f}")
    results[name] = {'Accuracy': accuracy}

    # 1c. Generate and display confusion matrix
    cm = confusion_matrix(y_val, y_pred)
    print("Confusion Matrix:")
    print(cm)
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix for {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # 2. Perform 5-fold cross-validation
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    print(f"Cross-validation accuracy (5-fold): {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")
    results[name]['CV_Mean_Accuracy'] = cv_scores.mean()
    results[name]['CV_Std_Accuracy'] = cv_scores.std()

print("\n--- Model Evaluation Completed ---")
```

## Select Best Model

### Subtask:
Automatically identify and select the machine learning model that achieved the highest validation accuracy during the evaluation phase. This model will be designated as the best for final predictions.


**Reasoning**:
I need to iterate through the evaluation results, compare the cross-validation mean accuracies, and identify the model with the highest accuracy as the best model, storing its name and the model object.



```python
best_model_name = ''
best_accuracy = 0
best_model = None

print("\n--- Selecting the Best Model ---")

for name, model_results in results.items():
    current_accuracy = model_results['CV_Mean_Accuracy']
    if current_accuracy > best_accuracy:
        best_accuracy = current_accuracy
        best_model_name = name
        best_model = models[name] # Retrieve the actual model object

print(f"The best model is '{best_model_name}' with a cross-validation mean accuracy of {best_accuracy:.4f}.")
print("Best model successfully selected and stored.")
```
